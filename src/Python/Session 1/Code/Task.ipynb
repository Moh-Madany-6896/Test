```python
email = "Amit_ml@gmail.edu"

x = email.find("@")
y = email.count("@")

if x != -1 and y == 1:
    username, domain = email.split("@")
    if "." in domain and domain.count(".") >= 1:
        print ("Valide")
        print (username)
        print (domain)
        if domain.endswith (".com"):
            print ("Commercial Domain")
        elif domain.endswith (".edu"):
            print ("Educational Domain")
        else:
            print ("Other Domain")
    else:
        print ("Invalide")
else:
    print ("Invalide")
```
