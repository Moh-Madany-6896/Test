```python
class mail_valid:
    """
    This Class check the Email Validity
    It Takes One Arg
        
        1st Arg >> email checked
    """
    def __init__(self):
        pass
    def validation (self, email):
        x = email.find("@")  # check if the @ is in the email given and not the last object
        y = email.count("@") # check the number of @ in the email
        if x != -1 and y == 1:
            username, domain = email.split("@")  # Divide the email to two parts
            if "." in domain and domain.count(".") >= 1:  # check the . presence and numbers in the email given
                print ("Valide")
                print (username)
                print (domain)
                if domain.endswith (".com"):     # Check the email type
                    print ("Commercial Domain")
                elif domain.endswith (".edu"):
                    print ("Educational Domain")
                else:
                    print ("Other Domain")
            else:
                print ("Invalide")
        else:
            print ("Invalide")


valid = mail_valid ()

valid.validation ("Amit_ml@gmail.edu")
```

```python
message = "###!!@mocleW EPGTQ!!!6789"

class decode:
    """ 
    This class used to decode the messages

    Arg: the coded text
    """
    def __init__(self):
        pass
    def Decode (self, text):
        characters = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
        result = []
        for x in text:   # extract the only characters from the message
            if x in characters or x == " ":
               result.append(x)
        
        result_2 = "".join(result)
        final = result_2 [::-1]
        return final


x = decode ()

x.Decode (message)
```

```python
class create_list:
    """
    Create a list of numbers that you want

    Arg: the number of items you want to enter in the list
    """
    def __init__(self, num):
        self.num = num
    def create (self):
        tries = self.num
        result = []
        while tries >= 1:
            number = int (input ("Enter Your Number, the tries left ({}): ".format (tries)))
            result.append (number)
            tries -= 1
        return result
    

x = create_list (5)

print (x.create())
```

```python
fruits = ['Apple', 'Banana', 'Orange']
class add ():
    """
    This Class check item if it present in a list or not
    
    if the item is present this class will remove it, if not present it'll be added
    
    Arg: the item that will be checked
    """
    def __init__(self):
        pass
    def Add_Fruit (self, fruit):
        f = fruit.capitalize ()
        if f in fruits:
            print ("This Fruit is Already Exists")
            fruits.remove (f)
        else:
            fruits.append (f)
        return fruits
    
x = add ()
x.Add_Fruit ("blueberry")
```

```python
class total:
    """
    This class makes any mathematical process to multible numbers

    Arg: 
       List of numbers or multible ny
    """
    def __init__(self):
        pass
    def T_Sum (self, *items):
        result = 0
        for x in items:
            result += x
        return result
    def T_Multi (self, *items):
        result = 1
        for x in items:
            result *= x
        return result
    

x = total ()

print (x.T_Sum (5, 10, 15, 20))
print (x.T_Multi (5, 10, 15, 20))
```

```python
class compare:
    def __init__(self):
        pass
    def Small (self, items):
        items.sort ()
        return items [0]
    def Large (self, items):
        items.sort ()
        return items [-1]


numbers = [10, 3, 25, 1, 7]
x = compare ()
print (x.Small (numbers))
print (x.Large (numbers))
```

```python
words = ["hi", "a", "hello", "no", "on"]

class count_num:
    def __init__(self):
        pass
    def num (self, item):
        result = 0
        for z in item:
            if len (z) >= 2:
                result += 1
        return result
    
x = count_num ()
x.num (words)
```

```python
data = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]

class change:
    def __init__(self):
        pass
    def change_last (self, item, value):
        result = []
        for z in item:
            convert = list (z)
            convert.remove (z[-1])
            convert.append (value)
            final = tuple (convert)
            result.append (final)
        return result
    
x = change ()
x.change_last (data, 100)
```

```python
values = (("item1", 10.5), ("item2", 5.7), ("item3", 8.9))

class Data:
    def __init__(self):
        pass
    def Data_Sort (self, items):
        result = sorted (items, key= lambda z : z[1])
        return result
    
x = Data ()
x.Data_Sort (values)
```

```python
words = ["hi", "hello", "fantastic", "world"]

class longest:
    def __init__(self):
        pass
    def Longest (self, items):
        result = sorted (items, key= lambda z : len (z))
        final = result [-1]
        return final

x = longest ()
x.Longest (words)
```

```python
data = {"a": 10, "b": 20, "c": 30}

class Total:
    def __init__(self):
        pass
    def value_total (self, items):
        result = 0
        for x in items.values ():
            result += x
        return result
    
x = Total ()
x.value_total (data)
```

```python
scores = {
    "student1": {"math": 80, "science": 90},
    "student2": {"math": 75, "science": 85},
    "student3": {"math": 92, "science": 88},
}

class Score:
    def __init__(self):
        pass
    def Student_Score (self, item):
        for a, b in item.items ():
            result = sum (b.values ())
            print (f"Total Score for {a} >> {result} Points")
    
x = Score ()
x.Student_Score (scores)
```
